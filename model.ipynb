```python
import pandas as pd
import numpy as np
```

```python
adults  = pd.read_csv("adult.data") #readingthedata
adults.info()
```

```python
#addingcollumnstothefile

adults.columns =['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'output'] 
adults.head(10)
```

```python
#random check for attribuites to find the missing value sign

print(adults['age'].unique(),
adults['workclass'].unique(),
adults['fnlwgt'].unique(),
adults['occupation'].unique(),
adults['race'].unique(),
adults['sex'].unique(),
adults['output'].unique())

#noted " ?" as the missed value
```

```python
#missingvlauecounts

l = len(adults.columns)

for i in range(l):
    to_check_col = adults.iloc[:,i]    
    null_list = to_check_col.isin([" ?"]).sum(axis = 0)
    print(adults.columns[i],"\t:",null_list)
```

```python
adults.replace(" ?", np.nan, inplace=True) #missingvalue " ?" replace with null count
adults.dropna(inplace=True) #dropping null
adults.isnull().sum() #confirming is there any missing values 
```

```python
adults.head(1)
```

```python
#set of attributes with catagorical values

attributes_set = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'output' ]
```

```python
#converting catogarical values to numerical 

for attribute in attributes_set:    
    labels = adults[attribute].astype('category').cat.categories.tolist() 
    replace_map_comp = { attribute : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}} 
    adults.replace(replace_map_comp, inplace= True)
```

```python
#confirm the transformation

print(adults['workclass'].unique(),
adults['education'].unique(),
adults['marital-status'].unique(),
adults['occupation'].unique(),
adults['race'].unique(),
adults['sex'].unique(),
adults['output'].unique())
```

```python
#preprocessing with minmax scaler

from sklearn import preprocessing
# ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'output'] 

scaler = preprocessing.MinMaxScaler()

adults[['age', 'workclass','fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'output']]= scaler.fit_transform(adults[['age', 'workclass','fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'output']])
```

```python
adults.head(10)
```

```python
#split data in to training and testing
from sklearn.model_selection import train_test_split

x = adults.drop(columns=['output'])
y = adults['output']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=35)
```

```python
adults.info()
```

```python
#checking the model

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt 


Bayes_clf = GaussianNB() #defeining the model
Bayes_clf.fit(x_train, y_train) #training
y_predict = Bayes_clf.predict(x_test) #testing


```

```python
accuracy = accuracy_score(y_test, y_predict)  #evaluating model accuracy 

print(accuracy) 
```

```python
#confusion Matrix 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
#distance to mean classifier

from sklearn.neighbors import NearestCentroid

#model defining
clf = NearestCentroid()
#train the model
clf.fit(x_train, y_train)
#testing the model with test data
y_predict = clf.predict(x_test)

#checking accuracy for evaluating the model 
accuracy = accuracy_score(y_test, y_predict)
print(accuracy)


```
